```python
# aivietnam.ai
# elu

import numpy as np

def elu_function(x):                                        
    return np.where(x >= 0, x, (np.exp(x) - 1)*alpha)

def elu_derivative(x, y):
    return np.where(x >= 0, 1, y+alpha)

alpha = 0.1
data = np.array([1,5,-4,3,-2])
print(elu_function(data))
```

```python
import matplotlib.pyplot as plt

x_axis = np.arange(-10,10,0.5)
y_axis = elu_function(x_axis)
plt.plot(x_axis, y_axis, color='g')
plt.xlabel('Giá trị input (x)', color=(255/255, 90/255, 17/255), fontsize=14)
plt.ylabel('Giá trị ELU (y)', color=(255/255, 90/255, 17/255), fontsize=14)

plt.tick_params(axis='x', colors='b')
plt.tick_params(axis='y', colors='b')
plt.show()
```

```python
# aivietnam.ai 
# elu function

import matplotlib.pyplot as plt
import numpy as np


# Calculate plot points
data = np.arange(-2., 2., 0.01)
data_elu  = elu_function(data)
data_delu = elu_derivative(data, data_elu)

# Setup centered axes
fig, ax = plt.subplots(figsize=(10, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Create and show plot
ax.plot(data, data_elu, color="#d35400", linewidth=3, label="elu")
ax.plot(data, data_delu, color="#1abd15", linewidth=3, label="derivative")
ax.legend(loc="upper left", frameon=False)
```

```python
#prelu
     
import numpy as np

alpha = 0.1
data = np.array([1,5,-4,3,-2])

def prelu_function(x):                                        
    return np.where(x > 0, x, x*alpha)

print(prelu_function(data))


import matplotlib.pyplot as plt

x_axis = np.arange(-10,10,0.5)
y_axis = prelu_function(x_axis)
plt.plot(x_axis, y_axis, color='g')
plt.xlabel('Giá trị input (x)', color=(255/255, 90/255, 17/255), fontsize=14)
plt.ylabel('Giá trị PReLU (y)', color=(255/255, 90/255, 17/255), fontsize=14)

plt.tick_params(axis='x', colors='b')
plt.tick_params(axis='y', colors='b')
plt.show()
```

```python
# aivietnam.ai 
# relu function
import matplotlib.pyplot as plt
import numpy as np

def relu_function(data):
    return (np.abs(data) + data) / 2

def relu_derivative(data):
    return (data > 0) * 1

# Calculate plot points
data = np.arange(-2., 2., 0.01)
data_relu  = relu_function(data)
data_drelu = relu_derivative(data_relu)

# Setup centered axes
fig, ax = plt.subplots(figsize=(10, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Create and show plot
ax.plot(data, data_relu, color="#d35400", linewidth=3, label="relu")
ax.plot(data, data_drelu, color="#1abd15", linewidth=3, label="derivative")
ax.legend(loc="upper left", frameon=False)
```

```python
# aivietnam.ai
# sigmoid - numpy

import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1./(1+np.e**(-x))

# Calculate plot points
x = np.arange(-9., 9., 0.01)
y = sigmoid(x)
dx = y*(1-y)

# Setup centered axes
fig, ax = plt.subplots(figsize=(10, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Create and show plot
ax.plot(x,y, color="#d35400", linewidth=3, label="sigmoid")
ax.plot(x,dx, color="#1abd15", linewidth=3, label="derivative")
ax.legend(loc="upper left", frameon=False)
```

```python
# aivietnam.ai
# tanh function

import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
    ex = np.exp(x)
    emx = np.exp(-x)
    return (ex - emx) / (ex + emx)

# Calculate plot points
x = np.arange(-5., 5., 0.1)
a = tanh(x)
dx = 1 - a**2

# Setup centered axes
fig, ax = plt.subplots(figsize=(10, 5))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Create and show plot
ax.plot(x, a, color="#d35400", linewidth=3, label="tanh")
ax.plot(x, dx, color="#1abd15", linewidth=3, label="derivative")
ax.legend(loc="upper left", frameon=False)
```
