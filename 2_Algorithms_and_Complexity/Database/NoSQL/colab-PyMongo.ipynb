##4.Aggregation Framework

```python
for i in test_db.student.aggregate([{"$match":{"name":"Thai"}}]):
    pprint.pprint(i)
```

```python
pipeline_agg = [
    {"$match":{"name":"Thai"}},
    {"$project":{"age":1}}
]
```

```python
for i in test_db.student.aggregate(pipeline_agg):
    pprint.pprint(i)
```
