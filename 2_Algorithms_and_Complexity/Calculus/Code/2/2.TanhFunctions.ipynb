```python
import matplotlib.pyplot as plt
import math

def plot_function_and_derivative(x, f_x, df_x):    
    # setting the axes at the centre
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # plot the function
    ax.plot(x, f_x, color="#d35400", linewidth=3, label='Function')
    ax.plot(x, df_x, color="#1abd15", linewidth=3, label='Derivative')
    
    ax.legend(loc="upper left", frameon=False)
    plt.show()
```

#### using derivative formula

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

def d_tanh(x):
    result = 1 - tanh(x)**2
    return result
```

```python
x_data = list(range(-100, 100+1, 1))
x_data = [x/10 for x in x_data]

y_data = [tanh(x) for x in x_data]
dy_data = [d_tanh(x) for x in x_data]

plot_function_and_derivative(x_data, y_data, dy_data)
```

#### using autograd

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

def autograd(x, epsilon=1e-6):    
    delta_y = tanh(x+epsilon) - tanh(x)
    derivative = delta_y/epsilon
    return derivative
```

```python
x_data = list(range(-100, 100+1, 1))
x_data = [x/10 for x in x_data]

y_data = [tanh(x) for x in x_data]
dy_data = [autograd(x) for x in x_data]

plot_function_and_derivative(x_data, y_data, dy_data)
```

#### autograd using numpy

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
import numpy as np

x_data = [x/10 for x in list(range(-100, 100+1))]
y_data = [tanh(x) for x in x_data]
dy_data = np.gradient(y_data, x_data).tolist()

plot_function_and_derivative(x_data, y_data, dy_data)
```
